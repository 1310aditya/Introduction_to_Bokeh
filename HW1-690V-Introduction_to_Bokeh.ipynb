```python
#Importing Libraries, Facilities and Setting Default Output
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
import pandas as pd
from bokeh.models import ColumnDataSource
from bokeh.layouts import column
from bokeh.models import HoverTool

#Output to Notebook
output_notebook()
```

```python
# Reading in the 'Alcohol consumption per adult 15+ (litres)' data using a URL
df = pd.read_excel("http://docs.google.com/spreadsheet/pub?key=0AgogXXPMARyldGJqTDRfNHBWODJMRWlZaVhNclhNZXc&output=xlsx")
```

```python
# Checking the head of table
df.head(3)
```

```python
# Storing Values of Index into a list
rows = df.index
rows = list(rows)
```

```python
# Plot Code Block
p1 = figure(plot_width = 10000, x_range = rows, toolbar_location = "left", toolbar_sticky = False)
p1.vbar(x = rows, width = 0.55, bottom = 0, top = df[2005], color="firebrick", legend = "Alcohol Consumption per adult in 2005")
p2 = figure(plot_width = 10000, x_range = rows, toolbar_location = "left", toolbar_sticky = False)
p2.vbar(x = rows, width = 0.55, bottom = 0, top = df[2008], color = "blue", legend = "Alcohol Consumption per adult in 2008")
p1.legend.location = "top_left"
p2.legend.location = "top_left"

# HoverTool Introduction
hover = HoverTool(tooltips=[
    ("Country", "@x"),
    ("Alcohol Consumption-2005", "@top Litres")
])

# Adding HoverTool to existing set of tools
p1.add_tools(hover)
p2.add_tools(hover)

# Displaying plots in single column
show(column(p1,p2))
```
